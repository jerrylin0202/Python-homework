```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interact
```

```python
x = np.linspace(-10,10,10000)
y = (x+9)*(x-2)*(x+8)*(x-7)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title(' Loss_Function ')
```

```python
def function(x):
    return (x+9)*(x-2)*(x+8)*(x-7)

def diffLossfunction(x):
    return 3*x**3+24*x**2-134*x-410

def gradient_descent(x_start,n,l_r):
    x_input = np.zeros(n+1)    
    x_input[0] = x_start
    for i in range(n):         
        ite = x_input[i]-l_r*(diffLossfunction(x_input[i]))
        x_input[i+1]=ite   
        
    return x_input
```

```python
trynum = np.random.uniform(-5,5,1)
go = gradient_descent(trynum,1000,0.001)
plt.figure(figsize=(10,10))
plt.plot(x,function(x), color='k')  
plt.scatter(go,function(go),c=go,cmap ='hsv_r')  
```
