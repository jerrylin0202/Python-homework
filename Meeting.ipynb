## Pandas

1. pandas中的表格叫做DataFrame，常以`df`來做為命名

2. 一些實用的指令:


 * df.head() `顯示欄位名稱與前五筆資料`
 
 
 * df.loc[] `選取列資料`
 
 
 * df['欄位名稱'].value `選取該欄的值`
 
 
 * df[0:456,'column名稱':'column名稱'] `選取要的表格範圍`
 
 
 * df.columns `變數資訊`
 
 
 * df.describe() `關於DataFrame各數值變數的描述統計`
 
 
 * df.index `這個DataFrame的列索引資訊`
 
 
 * df.info() `關於DataFrame的詳細資訊`
 
 
 * df(df['欄位名稱']=='欄位中的內容') 選出該欄中有包含該字元的欄位
 
 甚至可以df((df['欄位名稱']=='欄位中的內容') |或& (df['欄位名稱']=='欄位中的內容')) 可以透過運算子使用or或and

## Github

1. 辦帳號，下載git bash
[Download](https://git-scm.com/)

2. 在你想要和github連結的資料夾中建立一個README.md，裡面內容隨便打沒關係

3. ![github starter](githubmeeting.jpg)

4. 所有初始化設定完成後，每次的動作就是

  (1).  cd 原本設定的資料夾

  (2).  git add .

  (3).  git commit -m "這次更新的重點"

  (4).  git push 你的repository 或是 git push -u origin master

 * origin的部分是可以更改的，但一般都傾向不去更改
 * 可以使用git pull 你的repository來把現在存在github上的東西下載到你的電腦
